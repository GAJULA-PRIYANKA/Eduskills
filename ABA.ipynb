### **29. How do you calculate the joint probability distribution?**
For **discrete random variables**, the joint probability distribution is calculated as:  
\[
P(X=x, Y=y) = \frac{\text{Number of times }(X=x, Y=y)\text{ occurs}}{\text{Total number of observations}}
\]  
For **continuous random variables**, it is defined using a joint probability density function \(f(x,y)\), where probabilities are found by integrating over regions:  
\[
P(a \leq X \leq b, c \leq Y \leq d) = \int_a^b \int_c^d f(x,y) \, dy \, dx
\]

### **30. What is the difference between a joint probability distribution and a marginal probability distribution?**
- **Joint distribution:** Describes the probability of two (or more) random variables occurring together.  
- **Marginal distribution:** Describes the probability of one variable alone, obtained by summing (discrete) or integrating (continuous) over the other variable(s).  
\[
P(X=x) = \sum_y P(X=x, Y=y) \quad \text{or} \quad P(X=x) = \int f(x,y) \, dy
\]

### **31. What is the covariance of a joint probability distribution?**
Covariance measures how two random variables vary together.  
\[
\text{Cov}(X,Y) = E[(X - E[X])(Y - E[Y])]
\]  
- Positive covariance → variables increase together.  
- Negative covariance → one increases while the other decreases.  
- Zero covariance → no linear relationship.

### **32. How do you determine if two random variables are independent based on their joint probability distribution?**

\]  
for all values of \(x\) and \(y\).  
In continuous cases:  
\[
f(x,y) = f_X(x) \cdot f_Y(y)
