# 2. What is the percent share of different property types and room types?

## üè† Percent Share of Property Types & Room Types  

### üîë Steps to Solve
1. **Group data by property type** (e.g., Apartment, House, Condo, Villa).  
2. **Group data by room type** (e.g., Entire home/apt, Private room, Shared room).  
3. **Count listings** in each category.  
4. **Calculate percent share**:  
   \[
   \text{Percent Share} = \frac{\text{Listings in Category}}{\text{Total Listings}} \times 100
   \]  
5. **Visualize**:  
   - Pie chart for property type distribution.  
   - Bar chart for room type distribution.

## üîë Why It‚Äôs Needed  
- Raw datasets often contain **long descriptive text** (e.g., ‚ÄúA cozy 2-bedroom apartment near the river with modern amenities and free Wi-Fi‚Äù).  
- Such text is **hard to visualize directly**.  
- By extracting categorical features, we make the data **structured, filterable, and dashboard-ready**.  

---

## üõ†Ô∏è Workflow

# Tableau Desktop/Tableau Public/PowerBI

# üõ†Ô∏è Using "Group and Replace" in Tableau / Power BI Prep  

## üîë Workflow  
1. **Load dataset** into Tableau Prep or Power BI Prep.  
2. **Select the Neighbourhood column**.  
3. **Apply ‚ÄúGroup and Replace‚Äù**:  
   - **By Pronunciation** ‚Üí Groups values that sound alike (e.g., ‚ÄúUniv Dist‚Äù vs. ‚ÄúUniversity District‚Äù).  
   - **By Spelling** ‚Üí Groups values with similar spelling (e.g., ‚ÄúMidtown‚Äù vs. ‚ÄúMid town‚Äù).  
   - **By Casing** ‚Üí Groups values differing only in letter case (e.g., ‚ÄúDowntown‚Äù vs. ‚ÄúDOWNTOWN‚Äù).  
4. **Review suggested groups** ‚Üí Manually confirm or adjust replacements.  
5. **Replace with standardized names** ‚Üí e.g., all variations of ‚ÄúDowntown‚Äù become ‚ÄúDowntown‚Äù.  
6. **Validate** ‚Üí Check counts to ensure listings are now aggregated correctly.
